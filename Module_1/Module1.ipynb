# 1.4

**calculate the monthwise VWAP**

```
df3 = df2.copy()
```

```
df3.info()
```

```
df3['year'] = df3.Date.dt.year
```

```
df3['month'] = df3.Date.dt.month
```

```
df4 = df3.copy()
```

```

def vwap(df):
    q = df['Total Traded Quantity'].values
    p = df['Close Price'].values
    return df.assign(vwap=(p * q).cumsum() / q.cumsum())


df4=df4.groupby([(df4.year),(df4.month)], group_keys=False).apply(vwap)
```

# 1.5

**calculate the average price over the last N days of the stock price data where N is a user defined parameter**

```
def avg_prive(N):
  df_n_days = df4.tail(N)
  print("Average Price Of last {} days is : {}".format(N,df_n_days['Close Price'].mean()))
  return
```

**calculate the profit/loss percentage over the last N days**

```

def profit_loss_pct(N):
    df_n_days = df4.tail(N)    
    
    profit_loss_prcnt = (df_n_days["Prev Close"] - df_n_days["Close Price"])/df_n_days["Close Price"]

    prof_loss = sum(profit_loss_prcnt)

    print("\nLast {} days Profit/Loss Result : \n ---------------- \n".format(N))

    if prof_loss < 0:
        print("Loss : ", prof_loss*100,"%")

    else:
        print("Profit : ", prof_loss*100,"%")

    return
```

Calculate the average price over the course of last - 1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year

```
avg_prive(7)
avg_prive(14)
avg_prive(30)
avg_prive(90)
avg_prive(180)
avg_prive(365)
```

# 1.7

**Add another column 'Trend'**

```
df6 = df5.copy()
```

```
cond = []

for i in df6['PCT_change']:
  if(i<=0.5 and i>=-0.5):
    cond.append("Slight or No change")
  elif(i>0.5 and i<=1):
    cond.append("Slight positive")
  elif(i<-0.5 and i>=-1):
    cond.append("Slight negative")
  elif(i>1 and i<=3):
    cond.append("Positive")
  elif(i<-1 and i>=-3):
    cond.append("Negative")
  elif(i>3 and i<=7):
    cond.append('Among top gainers')
  elif(i<-3 and i>=-7):
    cond.append('Among top losers')
  elif(i>7):
    cond.append("Bull run")
  elif(i<-7):
    cond.append("Bear drop")

print(cond)
```

```
print(len(cond))
```

```
df6["Trend"] = pd.Series(cond, index=df6.index)
```

```
df6.head(10)
```
